```python
#rain itu akumulatif
#rain =rainnc+rainc

rn=np.empty((25,3,2)) #define array rain accumulation, 25=time, 3=lat, 2=lon
for i in range(0,25):
    rn[i]=file['rainc'][i,0,53:56,64:66]+file['rainnc'][i,0,53:56,64:66] #sum rain #get data for [time i, lev 0, lat jaktim, lon jaktim]
    print(i)
    
rn_new=np.empty((25,3,2)) #define array rain per time 
rn_new[0]=rn[0]
for i in range(0,24):
    rn_new[i+1]=rn[i+1]-rn[i] #accumulate, (rain+1)-rain
    print(i+1)
```
